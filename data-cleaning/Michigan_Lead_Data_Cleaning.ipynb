```python
import pandas as pd
```

```python
history_data = pd.read_excel(
    "90th percentiles over time.xlsx",
    sheet_name="2016_2025 Duplicates Removed"
)

history_data.shape
history_data.head()
```

```python
history_data.head()
```

```python
history_data.info()
```

```python
history_data = history_data.iloc[:, 0:8]

history_data.shape
history_data.head()
```

```python
history_data.shape
```

```python
history_data.columns
```

```python
history_data.columns = history_data.columns.str.replace("\n", " ", regex=False)
history_data.columns = history_data.columns.str.strip()
history_data.columns = history_data.columns.str.lower()
history_data.columns = history_data.columns.str.replace(" ", "_")
history_data.columns = history_data.columns.str.replace("(", "", regex=False)
history_data.columns = history_data.columns.str.replace(")", "", regex=False)

history_data.columns
```

```python
history_data = history_data.rename(columns={
    "public_water_supply_id": "pwsid",
    "lead_90th_percentile_ppb": "lead_90th_ppb",
    "last_monitoring_period_end": "monitoring_end_date",
    "includes_5th_liter?": "includes_5th_liter_or_not"
})
```

```python
history_data["monitoring_end_date"] = pd.to_datetime(history_data["monitoring_end_date"])

history_data["year"] = history_data["monitoring_end_date"].dt.year
history_data["month"] = history_data["monitoring_end_date"].dt.month
```

```python
history_data["lead_90th_ppb"] = pd.to_numeric(history_data["lead_90th_ppb"], errors="coerce")
```

```python
print("Shape (rows, cols):", history_data.shape)
```

```python
print("Columns:")
print(list(history_data.columns))
```

```python
key_cols = ["pwsid", "system_name", "monitoring_end_date", "lead_90th_ppb"]
print("Missing values in key columns:")
print(history_data[key_cols].isna().sum())
```

```python
display(history_data.head(10))
```

```python
print(len(history_data))
print(history_data.isna().all(axis=1).sum())
```

```python
display(history_data[history_data["pwsid"].isna()].head(10))
```

```python
history_data = history_data.dropna(
    subset=["pwsid", "monitoring_end_date", "lead_90th_ppb"]
)

print(history_data.shape)
```

```python
print("Final shape:", history_data.shape)
print("Unique systems:", history_data["pwsid"].nunique())
print("Year range:",
      history_data["year"].min(),
      "-",
      history_data["year"].max())
```
