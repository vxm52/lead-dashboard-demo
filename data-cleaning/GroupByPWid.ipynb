```python
history_clean = history_data.copy()
```

```python
history_clean["monitoring_end_date"] = pd.to_datetime(
    history_clean["monitoring_end_date"]
)
```

```python
duplicate_checking = history_clean.copy()

duplicate_checking["lead_value_count"] = (
    duplicate_checking
    .groupby(["pwsid", "monitoring_end_date"])["lead_90th_ppb"]
    .transform("nunique")
)
```

```python
duplicate_checking["system_name_count"] = (
    duplicate_checking
    .groupby(["pwsid", "monitoring_end_date"])["system_name"]
    .transform("nunique")
)
```

```python
need_review = duplicate_checking[
    duplicate_checking["lead_value_count"] > 1
]
```

```python
need_review = need_review.sort_values(
    ["pwsid", "monitoring_end_date", "system_name"]
)
```

```python
history_clean["above_action_level"] = (
    history_clean["lead_90th_ppb"] > 12
)
```

```python
need_review.to_csv(
    "lead_90th_history_need_review.csv",
    index=False
)
```
